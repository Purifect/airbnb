## Step 2 work on Bathroom Texts
___

### Create list of unique values in 'bathroom_texts'

```python
bathroom_texts_list = listings_reduced['bathrooms_text'].unique()

print(bathroom_texts_list)
```

### Make nan into string -> 'NA'

In order for the the dictionary of terms I'm going to create it I found it necessary to make NA into a string 'NA'

Later I will change them back

```python
#convert NA values to 'NA'
listings_reduced['bathrooms_text'] = listings_reduced['bathrooms_text'].astype(object).replace(np.nan, 'NA')
```

### Create dictionary of terms from Unique values 


In this step I doing several things. 
1) the NA term must string -> 'NA' \
2) half bath /  private half bath entries now have a numeric lead of 0.5 (even if its still a string value) \
3) I make each entry into private or shared 
>* If bath appears alone I assume it to be private
 

```python
transdict = {'1 bath': '1 private', 
             'NA': 'NA',
             '1.5 baths': '1.5 private',
             '1 shared bath': '1 shared', 
             '1 private bath': '1 private',
             'Shared half-bath': '0.5 shared',
             '2 baths': '2 private',
             '1.5 shared baths': '1.5 shared',
             '3 baths': '3 private',
             'Half-bath': '0.5 private',
             '2.5 baths': '2.5 private', 
             '2 shared baths': '2 shared',
             '0 baths' : '0 private',
             '4 baths':'4 private', 
             '0 shared baths': '0 shared',
             'Private half-bath': '0.5 private',  
             '4.5 baths' :'4.5 private', 
             '5.5 baths': '5.5 private' , 
             '2.5 shared baths': '2.5 shared',
             '3.5 baths': '3.5 private', 
             '15.5 baths': '15.5 private', 
             '5 baths': '5 private', 
             '3 shared baths': '3 shared', 
             '4 shared baths': '4 shared',
             '6 baths': '6 private', 
             '3.5 shared baths': '3.5 private', 
             '6 shared baths' : '6 shared',
             '6.5 shared baths': '6.5 shared', 
             '4.5 shared baths' :'4.5 private',
             '5.5 shared baths' : '5.5 private', 
             '7.5 baths': '7.5 private', 
             '5 shared baths': '5 shared',
             '6.5 baths': '6.5 private', 
             '8 baths': '8 private', 
             '7 shared baths':'7 shared', 
             '7 baths': '7 private', 
             '10 baths': '10 private'}
```

```python
#use trandict to iterate over bathrooms_text column

listings_reduced['bathrooms_text'] = [transdict[letter] for letter in listings_reduced['bathrooms_text']]
```

```python
# use str.split to create two new columns 'num_baths' and 'bath_type'
listings_reduced[['num_baths','bath_type']] = listings_reduced['bathrooms_text'].str.split(' ', 1, expand = True)
```

```python
#check work
listings_reduced.head(100)
```

___
### Step 3: Force num_baths to become float
___

```python
#convert 'NA' values to nan
listings_reduced['num_baths'] = listings_reduced['num_baths'].astype(object).replace( 'NA',np.nan)
```

```python
listings_reduced['num_baths'] = listings_reduced['num_baths'].astype(np.float16)
```

___
### Step: 4 Force bath_type to become category
___

```python
#convert None values to nan
listings_reduced.bath_type.fillna(value=np.nan, inplace=True)
```

```python
listings_reduced['bath_type'] = listings_reduced['bath_type'].astype('category')
```
